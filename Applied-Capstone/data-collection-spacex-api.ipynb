```python
import requests
import pandas as pd
import datetime
```

```python
# URL da API da SpaceX para lançamentos passados
spacex_url = "https://api.spacexdata.com/w4/launches/past"
```

```python
# Fazendo a requisição GET para a API
response = requests.get(spacex_url)
```

```python
# Convertendo a resposta JSON para um dataframe
data = pd.json_normalize(response.json())
```

```python
# Selecionando apenas as colunas de interesse
data = data[['rocket', 'payloads', 'launchpad', 'cores', 'flight_number', 'date_utc']]

# Filtrando linhas com apenas um core (removendo foguetes com boosters extras)
data = data[data['cores'].map(len) == 1]

# Filtrando linhas com apenas um payload
data = data[data['payloads'].map(len) == 1]

# Extraindo o único valor das listas de cores e payloads
data['cores'] = data['cores'].map(lambda x: x[0])
data['payloads'] = data['payloads'].map(lambda x: x[0])

# Convertendo a data para formato datetime e extraindo apenas a data (sem hora)
data['date'] = pd.to_datetime(data['date_utc']).dt.date

# Filtrando lançamentos até 13 de novembro de 2020
data = data[data['date'] <= datetime.date(2020, 11, 13)]
```
